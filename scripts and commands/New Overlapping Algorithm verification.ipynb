```python
import numpy as np
from scipy.spatial import KDTree
from scipy.spatial.distance import pdist, squareform
from time import time
```

```python
points_count = 20000
spread = 50

def get_random_points(count, data_spread):
    return (np.random.rand(count, 2) - 0.5) * data_spread

points = get_random_points(points_count, spread)
overlapping_array = squareform(pdist(points) < 0.1)
overlapping_pairs = KDTree(points).query_pairs(0.1)

for i in range(200):
    points = get_random_points(points_count, spread)

    injected_particles = np.random.randint(points_count, size = 9)
    print(f"Injected Particles: {injected_particles}")

    overlapping_array[injected_particles, :] = True
    overlapping_array[:, injected_particles] = True
    overlapping_array[squareform(np.invert(pdist(points) < 0.1))] = False

    pairs = KDTree(points).query_pairs(0.1)

    for pair in pairs:
        i = pair[0]
        j = pair[1]

        in_array = overlapping_array[i, j]
        in_set = pair in overlapping_pairs or i in injected_particles or j in injected_particles

        if in_array != in_set:
            print(f"Mismatch {pair} Array: {in_array} Set: {in_set}")

        overlapping_array[i,j] = True
        overlapping_pairs.add(pair)

    overlapping_pairs = pairs

    overlapping_array[injected_particles, :] = False
    overlapping_array[:, injected_particles] = False
```
